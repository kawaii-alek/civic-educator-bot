```python
!pip install requests pandas numpy  # For API calls and data handling

import requests

def ask_ollama(question, model="deepseek-r1:1.5b"):
    url = "http://localhost:11434/api/generate"
    payload = {
        "model": model,
        "prompt": f"You are a Kenyan civic education expert. Answer concisely: {question}",
        "stream": False  # Set to True for streaming responses
    }
    response = requests.post(url, json=payload)
    if response.status_code == 200:
        return response.json()["response"]
    else:
        return f"Error: {response.text}"

# Example usage
question = "What does Article 26 say about the right to life?"
answer = ask_ollama(question)
print("Question:", question)
print("Answer:", answer)
```
